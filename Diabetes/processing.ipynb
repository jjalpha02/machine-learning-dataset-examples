```python
import pandas as pd 
from sklearn import linear_model
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

df = pd.read_csv('song_data.csv')

x = df['song_duration_ms'].values
y = df['song_popularity'].values 
print(f'Song Duration: {x}')
print(f'Song Popularity: {y}')

x = x.reshape(-1,1)
y = y.reshape(-1,1)

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.075, test_size=0.025)
# print(f'Song Duration Train Data shape{x_train.shape}')
# print(f'Song Duration Test Data shape{x_test.shape}')
# print(f'Song Popularity Train Data shape{y_train.shape}')
# print(f'Song Popularity Test Data shape{y_test.shape}')

# plt.rcParams['figure.figsize'] = [20, 12]
# plt.scatter(x_train, y_train, color='purple')
# plt.xlabel('Song\'s Duration (ms)')
# plt.ylabel('Song\'s Popularity')
# plt.title('Training Data')
# plt.show()

lm = LinearRegression()
lm.fit(x_train, y_train)
y_predict = lm.predict(x_test)
print(f'Train accuracy {round(lm.score(x_train,y_train)*100, 5)}%')
print(f'Test accuracy {round(lm.score(x_test,y_test)*100, 5)}%')

plt.scatter(x_train, y_train, color="green")
plt.plot(x_test, y_predict)
plt.xlabel("Song's Duration (ms)")
plt.ylabel("Song's Popularity")
plt.title('Training Data')
plt.show()

```
